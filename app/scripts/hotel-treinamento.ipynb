```python
colunas = []
colunas.append('label_avg_price_per_room')
for i in range(len(base_hotel.columns[:-1])):
    colunas.append(base_hotel.columns[i])
base_hotel = base_hotel[colunas]
base_hotel
```

```python
base_hotel = pd.get_dummies(base_hotel,prefix=['type_of_meal_plan', 'room_type_reserved', 'arrival_year', 'market_segment_type'],
                   columns=['type_of_meal_plan', 'room_type_reserved', 'arrival_year', 'market_segment_type'])
```

```python
base_hotel = base_hotel.astype(int)
base_hotel
```

# Treinamento Do XGBoost

```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
hotel_train = pd.read_csv('hotel_train_xgboost.csv')
```

```python
X_train = hotel_train.iloc[:, 1:].values  # Todas as colunas, exceto a primeira (que é a label)
```

```python
y_train = hotel_train.iloc[:, 0].values   # Primeira coluna como a label
```

```python
xgb_model = xgb.XGBClassifier(objective='multi:softmax', num_class=3)
```

```python
xgb_model.fit(X_train, y_train)
```

```python
y_pred = xgb_model.predict(X_train)
```

```python
print(f"Acurácia: {accuracy_score(y_train, y_pred)}")
```

```python
print("Matriz de confusão:")
print(confusion_matrix(y_train, y_pred))
```

```python
print("Relatório de classificação:")
print(classification_report(y_train, y_pred))
```
